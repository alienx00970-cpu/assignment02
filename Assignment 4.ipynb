### E-Commerce Product Dictionary

```python
# Question # 01
# An e-commerce store stores information about its products in a nested dictionary. The outer dictionary uses product IDs as keys, and the inner dictionary
# stores product details like name, category, price, and stock quantity. products = { 101: {"name": "Laptop", "category": "Electronics", "price": 1200,
# "stock": 50}, 102: {"name": "Shirt", "category": "Apparel", "price": 25, "stock": 200}, 103: {"name": "Coffee Maker", "category": "Home Appliances",
# "price": 80, "stock": 30} }
# • Increase shirt's stock from 200 to 250
# • Add a new product (e.g., "Smartphone")

# Expected Output:
# {101: {'name': 'Laptop', 'category': 'Electronics', 'price': 1200, 'stock': 50}, 102: {'name': 'Shirt', 'category': 'Apparel', 'price': 25,
# 'stock': 250}, 103: {'name': 'Coffee Maker', 'category': 'Home Appliances', 'price': 80, 'stock': 30}} {101: {'name': 'Laptop', 'category': 'Electronics',
# 'price': 1200, 'stock': 50}, 102: {'name': 'Shirt', 'category': 'Apparel', 'price': 25, 'stock': 250}, 103: {'name': 'Coffee Maker', 
# 'category': 'Home Appliances', 'price': 80, 'stock': 30}, 104: {'name': 'Smartphone', 'category': 'Electronics', 'price': 800, 'stock': 100}}

products = {
    101: {"name": "Laptop", "category": "Electronics", "price": 1200, "stock": 50},
    102: {"name": "Shirt", "category": "Apparel", "price": 25, "stock": 200},
    103: {"name": "Coffee Maker", "category": "Home Appliances", "price": 80, "stock": 30}
}

products[102]["stock"] = 250

print("Updated Products:")
print(products)

products.update({
    104: {
        "name": "Smartphone",
        "category": "Electronics",
        "price": 800,
        "stock": 100
    }
})

print("\nAfter Adding Smartphone:")
print(products)
```

### Set Difference and Symmetric Difference

```python
# Question # 04 
# You are managing a list of students enrolled in two different courses. You need to perform various set operations to understand the student enrollment.
# Task: Create two sets: one for students in "Course A" and one for students in "Course B". Find the students who are in "Course A" but not in
#  "Course B" (difference). Find students who are only in one of the two courses (symmetric difference). course_a = {"John", "Alice", "Bob", "David"} 
# course_b = {"Alice", "Eve", "Charlie", "David"}

# Using set methods
# • Find students who are in Course A but not in Course B (difference)
# • Find students who are only in one of the two courses (symmetric difference)

course_a = {"John", "Alice", "Bob", "David"}
course_b = {"Alice", "Eve", "Charlie", "David"}

difference_students = course_a.difference(course_b)

symmetric_students = course_a.symmetric_difference(course_b)

print(difference_students)

print(symmetric_students)
```

### Inventory Dictionary

```python
# Question # 07
#  inventory = { "The Great Gatsby": {"price": 10.99, "quantity": 5}, "Rich dad poor dad": {"price": 8.99, "quantity": 8},
# "Harry Potter": {"price": 12.50, "quantity": 2}, } • Add a new book in inventory using method and get all book titles using method 
# • Remove a book if it's out of stock

inventory = {
    "The Great Gatsby": {"price": 10.99, "quantity": 5},
    "Rich dad poor dad": {"price": 8.99, "quantity": 8},
    "Harry Potter": {"price": 12.50, "quantity": 2}
}
inventory.update({
    "The Iron Widow": {
        "price": 15.99,
        "quantity": 4
    }
})
print("Book Titles:")
print(inventory.keys())

inventory["Harry Potter"]["quantity"] = 0

if inventory["Harry Potter"]["quantity"] == 0:
    inventory.pop("Harry Potter")

print("\nUpdated Inventory:")
print(inventory)
```

### Combine Two Sets

```python
# Question # 09
# . Combine Two Sets A = {"Ali", "Sara"} B = {"Ahmed", "Zara"}
# Combine both sets into one set.

A = {"Ali", "Sara"}
B = {"Ahmed", "Zara"}
combined_set = A.union(B)
print(combined_set)
```

### Set Comparison

```python
# Question # 10
# Set Comparison A = {1, 2, 3, 4, 5} B = {4, 5, 6, 7}
# Find common elements. Find elements only in Set A. Find elements only in Set B.
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

common = A.intersection(B)
only_A = A.difference(B)
only_B = B.difference(A)

print("Common elements:", common)
print("Only in Set A:", only_A)
print("Only in Set B:", only_B)
```
