```python
#Q4 : Check the type and convert:\n","(num1 = 10, num2 = 5.6, num3 = 4.43, num4='23')
#Integer to float,Float to integer,Float to String, String to Integer

num1 = 10
num2 = 5.6
num3 = 4.43
num4 = '23'

print(type(num1))
print(type(num2))
print(type(num3))
print(type(num4))

int_to_float = float(num1)
float_to_int = int(num2)
float_to_string = str(num3)
string_to_int = int(num4)

print(int_to_float)
print(float_to_int)
print(float_to_string)
print(string_to_int)
```
