### Guess the Secret Word
### Question # 01

```python
# Question # 01
# Write a Python program that asks the user to guess a secret word. Keep asking until the user enters the correct word ("python").
# When guessed correctly, print "🎉 You guessed it!"
secret_word = "python"

while True:  # ← This is ALL you need!
    guess = input("Guess the word: ")
    if guess == secret_word:
        print("🎉 You guessed it!")
        break
    else:
        print("Try Again!")
```

###  Stop Sending Emails After 3 Users
### Question # 02

```python
# Question # 02
# Stop sending emails after 3 users (Using for loop and break)
users = ["Ali", "Sara", "John", "Zara", "Mike"]

for i, user in enumerate(users, 1):
    if i > 3:
        print("Limit reached. Stop sending emails.")
        break
    print(f"Email sent to {user}")
```

###  Stop Monitoring Temperature
### Question # 03

```python
# Question # 03
#  Stop monitoring temperature if critical temperature reached (Using for loop and break), temps = [22, 25, 30, 45, 28] critical = 40
temps = [22, 25, 30, 45, 28]
critical = 40

for temp in temps:
    print(f"Temperature: {temp}")
    if temp >= critical:
        print("Critical temperature reached! Stop monitoring.")
        break
```

### Calculate Total Price

```python
# Question # 04
# Calculate total price of items in a shopping cart (Using For loop)
cart = [100, 250, 50, 300]
total = 0

for price in cart:
    total += price
print(f"Total Price: {total}")
```

### Skip Customers with Unpaid Bills 
### Question # 05 

```python
# Question # 05
#  Skip customers with unpaid bills (Using For loop and continue)
customers = [
    {"name": "Ali", "paid": True},
    {"name": "Sara", "paid": False},
    {"name": "John", "paid": True}
]
for customer in customers:
    if customer["paid"] == True:
        print(f"Access granted to: {customer['name']}")
```

### Extract Domain Names
### Question # 07



```python
# Question # 07
# Extract domain names from email list (Using for loop)
emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]
domains = []
for email in emails:
    domain = email.split('@')[1]
    domains.append(domain)
print(f"domains: {domains}")
```

### Print only Overdue Tasks
### Question # 08

```python
# Question # 08
# Print only overdue tasks (Using for loop)
tasks = [
    {"task": "Pay bills", "status": "done"},
    {"task": "Submit report", "status": "pending"},
    {"task": "Clean house", "status": "pending"}
]

for task in tasks:
    if task["status"] == "pending":
        print(f"Overdue: {task['task']}")
```

### Stop on Negative Number
### Question # 09


```python
# Question # 09
# Write a Python program that keeps asking the user to enter numbers and prints them. The program should stop the loop if the user enters a negative
# number. Use a while True loop with a break statement
while True:
    num = int(input("Enter a number (negative to stop): "))
    if num < 0:
        print("Negative number entered. Loop stopped!")
        break
    print(f"You entered: {num}")
```

### Skip Weekends
### Question # 10

```python
# Question # 10
# Skip weekends while printing workdays (Using for loop and continue)
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for day in days:
    if day in ["Saturday", "Sunday"]:
        continue
    print(f"Workday: {day}")
```
