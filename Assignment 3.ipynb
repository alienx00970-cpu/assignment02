### Product Code

```python
# Question # 07
# Find the first position of the dash (-).Count the number of dashes (-) in the product code.
# Remove the "SKU-" prefix and "-XYZ" suffix from the product code, Extract the numeric part 12345
code = "SKU-12345-XYZ"

first_dash = code.find("-")
dash_count = code.count("-")
number = code.replace("SKU-", "").replace("-XYZ", "")

print("First dash position:", first_dash)
print("Number of dashes:", dash_count)
print("Numeric part:", number)
```

### String Processing

```python
# Question # 09
# Perform the following tasks:
text = " Data Analysis with Python is very powerful "

text = text.strip()
title_text = text.title()
replaced_text = title_text.replace("Powerful", "Important")
position = replaced_text.find("Python")

print("Title Case:", title_text)
print("After Replace:", replaced_text)
print("Position of Python:", position)
```
